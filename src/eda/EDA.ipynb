### Data Cleaning and Fixing Anomalies
From our profiling, we notice:
- Zeros and extreme values in `price`.
- `last_review` column is string format instead of datetime.

We will filter prices between $10 and $350, and convert `last_review` to datetime.

```python
# Drop outliers
min_price = 10
max_price = 350
idx = df['price'].between(min_price, max_price)
df = df[idx].copy()

# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])

# Inspect changes
df.info()
```
